```python
import pandas as pd
import requests
import json
```

```python
api_route_raw = open('API_SETTINGS.json')
api_route = json.loads(api_route_raw.read())
```

```python
api_route
```

```python
twitstats_base=api_route['API_BASE_ROUTE']
```

```python
data = requests.get(twitstats_base+'posts/user/NyanNyanners/30')\
                         .json()
df = pd.DataFrame.from_dict(data)
```

```python
data_set2 = requests.get(twitstats_base+'posts/user/elonmusk/30')\
                         .json()
```

```python
combined_set=pd.concat([df,pd.DataFrame.from_dict(data_set2)])
```

```python
combined_set.loc[(combined_set['media'] == "text") & (combined_set['favorites'] >= 10000)]
```

```python
df_local_file1 = pd.read_json("jiyuna_10212021_last30d.json")
```

```python
df_local_file1  = df_local_file1.drop(['text'], axis=1)
```

```python
df_reduced = df_local_file1.loc[(df_local_file1['favorites'] >= 1000)]
combined_set=pd.concat([combined_set, df_reduced])
```

```python
hashtags_set=combined_set[~combined_set['hashtags'].isnull() & combined_set['hashtags'].str.len() > 0]
```

```python
hashtags_set.to_csv('twitter_resutls_with_hashtag_10212021.csv')
```

```python
parsed = json.loads(hashtags_set.drop(['text'], axis=1).to_json(orient="records"))

with open('data.json', 'w') as outfile:
     json.dump(parsed , outfile, sort_keys = True, indent = 4,
               ensure_ascii = False)
```
